```python
import pandas as pd
```

```python
Book_data = pd.read_csv("E:\\Desktop\\human activity\\Books.csv")
Book_data
```

```python
User_data = pd.read_csv("C:\\Users\\HP\\Desktop\\recover\\Users.csv")
User_data
```

```python
Book_Rating = pd.read_csv("E:\\Desktop\\human activity\\Ratings.csv")
Book_Rating
```

```python
Book=Book_data.merge(Book_Rating,on='ISBN')[["Book-Title", "Book-Author", "Year-Of-Publication","Publisher","Image-URL-S","User-ID","Book-Rating"]]
Book.head(10)
```

```python
Book.isnull().sum()
```

```python
Book.duplicated().sum()
```

```python
df1=Book.groupby("Book-Title")["Book-Rating"].count().reset_index()
df1.rename(columns={"Book-Rating":"Rating_count"},inplace=True) 

```

```python
df2=Book.groupby("Book-Title")["Book-Rating"].mean().reset_index()
df2.rename(columns={"Book-Rating":"Avg_Rating"},inplace=True) 
```

```python
joint_df=df1.merge(df2,on="Book-Title")
Top_50_Books=joint_df[(joint_df["Rating_count"]>250) & (joint_df["Avg_Rating"]>3.5)].sort_values("Rating_count",ascending=False).head(50).reset_index(drop=True)
Top_50_Books=Top_50_Books.merge(Book,on="Book-Title").drop_duplicates(subset=["Book-Title"])[["Book-Title", "Book-Author", "Year-Of-Publication","Image-URL-S"]].reset_index(drop=True)

Top_50_Books.index=Top_50_Books.index+1
Top_50_Books
```

```python
#collabertive base recommendation system
#filtering data frame based on user, who have rated more than 50 books
active_book_reader=Book.groupby("User-ID")["Book-Rating"].count()>=50
active_book_reader_uid=active_book_reader[active_book_reader].index
Book[Book["User-ID"].isin(active_book_reader_uid)]
```

```python
#filtering data frame based on Book, which got 250+ ratings
famous_book=Book.groupby("Book-Title")["Book-Rating"].count()>=200
famous_book=famous_book[famous_book].index
final_data_frame=Book[Book["Book-Title"].isin(famous_book)]
final_data_frame2=final_data_frame.drop_duplicates(subset=["Book-Title"],keep="first")
index=final_data_frame2["Book-Title"]
final_data_frame2.index=index
final_data_frame
```

```python
# converting data frame into pivot table with colums as user_id and index as Book-title
Book_user_pt=final_data_frame.pivot_table(index="Book-Title",columns="User-ID",values="Book-Rating").fillna(0)
column=Book_user_pt.index
column
```

```python
Book_user_pt.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cs=cosine_similarity(Book_user_pt)
Book_similarity_table=pd.DataFrame(cs,columns=column,index=column)
Book_similarity_table
```

```python
# function to find top 5 books based on reading patterns by users 
from collections import Counter
class Book_search:
    @staticmethod
    def similarity_search(book_name,data,data2):
        list=[]
        top_5_similar_book=(data[book_name].sort_values(ascending=False).head(6)[1:].index)
        
        for x in top_5_similar_book:
            
            list.append(data2.loc[x].values)
        
        return list
BS=Book_search()
BS.similarity_search("A Case of Need",Book_similarity_table,final_data_frame2)
    
```
